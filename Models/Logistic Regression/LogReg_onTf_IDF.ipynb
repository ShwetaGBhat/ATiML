```python
import warnings
warnings.filterwarnings("ignore")
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import classification_report, confusion_matrix
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import OneHotEncoder 
from sklearn.compose import ColumnTransformer 
```

```python
data = pd.read_csv(r"/Users/surabhi/Desktop/ATML_Project/InitialDataPrep-FeatureExtractionPhase-May-2020/DataPreprocessed/Preprocessed_stopword.csv")
```

```python
data.head(3)
```

```python
enc1 = OneHotEncoder(handle_unknown='ignore')
enc_df1 = pd.DataFrame(enc1.fit_transform(data[['Book_Name']]).toarray())
 # merge with main df bridge_df on key values
del data['Book_Name']
col_name=[]
for i in range(enc_df1.columns.start,enc_df1.columns.stop,enc_df1.columns.step):
    x="Book_Name"+(str(i))
    col_name.append(x)
enc_df1.columns=col_name
data = data.join(enc_df1)
```

```python
enc1 = OneHotEncoder(handle_unknown='ignore')
enc_df1 = pd.DataFrame(enc1.fit_transform(data[['Author_Name']]).toarray())
 # merge with main df bridge_df on key values
del data['Author_Name']
col_name=[]
for i in range(enc_df1.columns.start,enc_df1.columns.stop,enc_df1.columns.step):
    x="Author_Name"+(str(i))
    col_name.append(x)
enc_df1.columns=col_name
data = data.join(enc_df1)
```

```python
alldata_train, alldata_test = train_test_split(data, test_size=0.3, random_state=42, stratify = np.array(data['guten_genre']))
```

```python
alldata_train.head(3)
```

```python
y_train = alldata_train['guten_genre']
y_test = alldata_test['guten_genre']
```

```python
del alldata_train['guten_genre']
del alldata_test['guten_genre']
```

```python
vectorizer = TfidfVectorizer(sublinear_tf=True)
X_train = vectorizer.fit_transform(alldata_train['clean_data'])
```

```python
X_test = vectorizer.transform(alldata_test['clean_data'])
```

```python
logreg = LogisticRegression(solver = 'lbfgs')
logreg.fit(X_train, y_train)
```

```python
y_pred = logreg.predict(X_test)
```

```python
count_misclassified = (y_test != y_pred).sum()
```

```python
from sklearn.metrics import accuracy_score, f1_score
from sklearn.metrics import classification_report, confusion_matrix

print('Testing accuracy %s' % accuracy_score(y_test, y_pred))
print('Testing F1 score: {}'.format(f1_score(y_test, y_pred, average='weighted')))
```

```python
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
print('Misclassification Count :' ,count_misclassified)
```

```python
from sklearn.feature_selection import chi2, SelectKBest
from sklearn.metrics import classification_report, confusion_matrix
from sklearn import metrics
from sklearn.naive_bayes import MultinomialNB
```

```python
k = [10000, 25000, 50000, 75000, 100000]
clf = LogisticRegression(solver = 'lbfgs')
ch, inferences, trained_model, accu = [], [], [], []
x_feat_train, x_feat_test = [], []

for i in range(0, len(k)):
    ch.append(SelectKBest(chi2, k[i]))
    x_feat_train.append(ch[i].fit_transform(X_train, y_train))
    x_feat_test.append(ch[i].transform(X_test))
    trained_model.append(clf.fit(x_feat_train[i], y_train))
    inferences.append(trained_model[i].predict((x_feat_test[i])))
    accu.append(metrics.f1_score(y_test, inferences[i], average='macro'))
    print('For K value:', k[i],'The F1 Score is',accu[i])
```

```python
k = [10000, 25000, 50000, 75000, 100000]
clf = MultinomialNB(alpha=0.001)
ch, inferences, trained_model, accu = [], [], [], []
x_feat_train, x_feat_test = [], []

for i in range(0, len(k)):
    ch.append(SelectKBest(chi2, k[i]))
    x_feat_train.append(ch[i].fit_transform(X_train, y_train))
    x_feat_test.append(ch[i].transform(X_test))
    trained_model.append(clf.fit(x_feat_train[i], y_train))
    inferences.append(trained_model[i].predict((x_feat_test[i])))
    accu.append(metrics.f1_score(y_test, inferences[i], average='macro'))
    print('For K value:', k[i],'The F1 Score is',accu[i])
```
