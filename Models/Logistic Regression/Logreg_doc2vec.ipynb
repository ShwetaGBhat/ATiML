```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
import numpy as np
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings("ignore")
```

```python
doc2vec_train = pd.read_csv(r"/Users/surabhi/Desktop/ATML_Project/InitialDataPrep-FeatureExtractionPhase-May-2020/DataPreprocessed/Train_doc2vec.csv")
```

```python
doc2vec_test = pd.read_csv(r"/Users/surabhi/Desktop/ATML_Project/InitialDataPrep-FeatureExtractionPhase-May-2020/DataPreprocessed/Test_doc2vec.csv")
```

```python
from sklearn.preprocessing import OneHotEncoder 
from sklearn.compose import ColumnTransformer 
```

```python
data = pd.concat ([doc2vec_train, doc2vec_test])
```

```python
enc1 = OneHotEncoder(handle_unknown='ignore')
enc_df1 = pd.DataFrame(enc1.fit_transform(data[['Book_Name']]).toarray())
 # merge with main df bridge_df on key values
del data['Book_Name']
col_name=[]
for i in range(enc_df1.columns.start,enc_df1.columns.stop,enc_df1.columns.step):
    x="Book_Name"+(str(i))
    col_name.append(x)
enc_df1.columns=col_name
data = data.join(enc_df1)
```

```python
enc1 = OneHotEncoder(handle_unknown='ignore')
enc_df1 = pd.DataFrame(enc1.fit_transform(data[['Author_Name']]).toarray())
 # merge with main df bridge_df on key values
del data['Author_Name']
col_name=[]
for i in range(enc_df1.columns.start,enc_df1.columns.stop,enc_df1.columns.step):
    x="Author_Name"+(str(i))
    col_name.append(x)
enc_df1.columns=col_name
data = data.join(enc_df1)
```

```python
data.head(3)
```

```python
doc2vec_train = data[0:695]
len(doc2vec_train)
```

```python
doc2vec_test = data[696:]
len(doc2vec_test)
```

```python
del doc2vec_train['Content']
del doc2vec_train['clean_data']
del doc2vec_train['id']
del doc2vec_train['cnt']
```

```python
del doc2vec_train['index']
```

```python
del doc2vec_test['Content']
del doc2vec_test['id']
del doc2vec_test['clean_data']
del doc2vec_test['cnt']
del doc2vec_test['index']
```

```python
y_train = doc2vec_train['guten_genre']
y_test = doc2vec_test['guten_genre']
```

```python
del doc2vec_train['guten_genre']
del doc2vec_test['guten_genre']
```

```python
logreg = LogisticRegression(solver = 'lbfgs')
logreg.fit(doc2vec_train, y_train)
```

```python
y_pred = logreg.predict(doc2vec_test)
```

```python
count_misclassified = (y_test != y_pred).sum()
```

```python
from sklearn.metrics import accuracy_score, f1_score
from sklearn.metrics import classification_report, confusion_matrix

print('Testing accuracy %s' % accuracy_score(y_test, y_pred))
print('Testing F1 score: {}'.format(f1_score(y_test, y_pred, average='weighted')))
```

```python
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
print('Misclassification Count :' ,count_misclassified)
```
