```python
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings("ignore")
from sklearn.linear_model import LogisticRegression
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder 
from sklearn.compose import ColumnTransformer 
```

```python
data = pd.read_csv(r"/Users/surabhi/Desktop/ATML_Project/InitialDataPrep-FeatureExtractionPhase-May-2020/DataPreprocessed/books_ALLFeatures.csv")
```

```python
data.head(3)
```

```python
del data['flesh_reading_score']
```

```python
enc1 = OneHotEncoder(handle_unknown='ignore')
enc_df1 = pd.DataFrame(enc1.fit_transform(data[['Book_Name']]).toarray())
 # merge with main df bridge_df on key values
del data['Book_Name']
col_name=[]
for i in range(enc_df1.columns.start,enc_df1.columns.stop,enc_df1.columns.step):
    x="Book_Name"+(str(i))
    col_name.append(x)
enc_df1.columns=col_name
data = data.join(enc_df1)
```

```python
enc1 = OneHotEncoder(handle_unknown='ignore')
enc_df1 = pd.DataFrame(enc1.fit_transform(data[['Author_Name']]).toarray())
 # merge with main df bridge_df on key values
del data['Author_Name']
col_name=[]
for i in range(enc_df1.columns.start,enc_df1.columns.stop,enc_df1.columns.step):
    x="Author_Name"+(str(i))
    col_name.append(x)
enc_df1.columns=col_name
data = data.join(enc_df1)
```

```python
data.head(3)
```

```python
alldata_train, alldata_test = train_test_split(data, test_size=0.3, random_state=42, stratify = np.array(data['guten_genre']))
```

```python
len(alldata_train)
```

```python
y_train = alldata_train['guten_genre']
```

```python
y_test = alldata_test['guten_genre']
```

```python
del alldata_train['guten_genre']
del alldata_test['guten_genre']
```

```python
del alldata_train['book_id']
del alldata_train['file']
```

```python
del alldata_train['clean_data']
```

```python
del alldata_train['Content']
```

```python
alldata_train.head()
```

```python
del alldata_test['book_id']
del alldata_test['file']
del alldata_test['clean_data']
```

```python
del alldata_test['Content']
```

```python
alldata_test.head()
```

```python
logreg = LogisticRegression(solver = 'lbfgs',penalty = 'l2', multi_class = 'multinomial')
logreg.fit(alldata_train, y_train)
```

```python
y_pred = logreg.predict(alldata_test)
```

```python
count_misclassified = (y_test != y_pred).sum()
```

```python
from sklearn.metrics import accuracy_score, f1_score
from sklearn.metrics import classification_report, confusion_matrix

print('Testing accuracy %s' % accuracy_score(y_test, y_pred))
print('Testing F1 score: {}'.format(f1_score(y_test, y_pred, average='weighted')))
```

```python
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
